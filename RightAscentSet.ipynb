This generates the right descent set for a given word read right to left. The algorithm is MUCH slower than the left ascent or descent

Pseudocode. The procedure has arguments s and the string s1 . . . sn representing w. It
uses variables
t: an element of S; initially s
i: in the range [0, n]; initially 0
k: also in the range [0, n]; initially 0
λ: an extended minimal root—either a minimal root or ⊕ or ⊖; initially αs
x: an element of W; initially 1
BEGINNING OF LOOP. At this point x = s1 . . . si−1, sx = s1 . . . sktsk+1si−1, λ = x
−1αs.
Read the next generator si
. Then x becomes x = xsi
. Set λ = siλ from the minimal root
reflection table.
If λ = β in ∆ with sβ < si
, then set t = sβ, k = i.
If λ = ⊖, this means k = i − 1 and si = t.
Set NF(sw) = s1 . . . si−1si+1 . . . sn and stop.
If λ = ⊕, set NF(sw) = s1 . . . sktsk+1sk+2 . . . sn and stop.
Advance i to i + 1. If i > n, stop with NF(sw) = s1 . . . sktsk+1sk+2 . . . sn. Else loop.
We now know how to use the minimal root reflection table to multiply, and it remains to
see how to construct it.

```python
def order(s):
    if s == 'a':
        return 0
    if s == 'b':
        return 1
    if s == 'c':
        return 2
    if s == 'd':
        return 3
    if s == 'e':
        return 4
    if s == 'f':
        return 5
def RootRefTable(s, a):
    Table = [['-', 'd', 'f'], ['d', '-', 'e'], ['f', 'e', '-'], ['b', 'a', '+'], ['+', 'c', 'b'], ['c', '+', 'a']]
    return Table[order(a)][order(s)]
def InsertChar(t, w, k):
    return (w[:k] + t + w[k:])
def RemoveChar(w, k):
    return (w[:k] + w[k+1:])
```

```python
def Mult(s, w):
    t = s
    k = 0
    λ = s
    for i in range(len(w)):
        #insert code
        λ = RootRefTable(w[i],λ)
        #print(w[i])
        #print(λ)
        if λ == '-':
            return RemoveChar(w, i)
        elif λ == '+':
            return InsertChar(t, w, k)
        elif order(λ) < order(w[i]):
            k = i+1
            t = λ
    return InsertChar(t, w, k)

def MultLen(s, w):
    λ = s
    for i in range(len(w)):
        λ = RootRefTable(w[i],λ)
        if λ == '-':
            return len(w)-1
        elif λ == '+':
            return len(w)+1
    return len(w)+1

def isRightAscent(s, w):
    w = NF(w[::-1])
    λ = s
    for i in range(len(w)):
        λ = RootRefTable(w[i],λ)
        if λ == '-':
            return False
        elif λ == '+':
            return True
    return True
```

```python
def NF(w):
    x = ''
    for i in range(1, len(w)+1):
        #print(x)
        x = Mult(w[-i], x)
    return x
```

```python
def rightAscent(w):
    w = w[::-1]
    list = ['a','b','c']
    returnlist = []
    for s in list:
        bool = MultLen(s,w) >= len(w)
        #print(m)
        #print(bool)
        if bool:
            returnlist.append(s)
    return returnlist
```

```python
def rightAscentPath(w):
    if len(w) == 1:
        gens = ['a', 'b', 'c']
        gens.remove(w)
        return [[w], gens]
    x = w[-1]
    elementList = [x]
    ascentsList = []
    for i in range(1, len(w)):
        gens = ['a', 'b', 'c']
        ascent = []
        for g in gens:
            #change to right
            if g == w[-i-1]:
                newx = Mult(g, x)
                elementList.append(newx)
            if isRightAscent(g, x):
                ascent.append(g)
        x = elementList[-1]
        #print(x)
        ascentsList.append(ascent)
    ascentsList.append(rightAscent(x))
    return [elementList, ascentsList]


```

```python
def order(s):
    if s == 'a':
        return 0
    if s == 'b':
        return 1
    if s == 'c':
        return 2
    if s == 'd':
        return 3
    if s == 'e':
        return 4
    if s == 'f':
        return 5
def RootRefTable(s, a):
    Table = [['-', 'd', 'f'], ['d', '-', 'e'], ['f', 'e', '-'], ['b', 'a', '+'], ['+', 'c', 'b'], ['c', '+', 'a']]
    return Table[order(a)][order(s)]
def InsertChar(t, w, k):
    if k == 0:
        return (t+w)
    else:
        return (w[:k] + t + w[k:])
def RemoveChar(w, k):
    if k == 0:
        return w[1:]
    else:
        return (w[:k] + w[k+1:])
```

```python
def Mult(s, w):
    t = s
    k = 0
    λ = s
    for i in range(len(w)):
        #insert code
        λ = RootRefTable(w[i],λ)
        #print(w[i])
        #print(λ)
        if λ == '-':
            return RemoveChar(w, i)
        elif λ == '+':
            return InsertChar(t, w, k)
        elif order(λ) < order(w[i]):
            k = i+1
            t = λ
    return InsertChar(t, w, k)

def MultLen(s, w):
    λ = s
    for i in range(len(w)):
        λ = RootRefTable(w[i],λ)
        if λ == '-':
            return len(w)-1
        elif λ == '+':
            return len(w)+1
    return len(w)+1

def isLeftDescent(s, w):
    λ = s
    for i in range(len(w)):
        λ = RootRefTable(w[i],λ)
        if λ == '-':
            return True
        elif λ == '+':
            return False
    return False
```

```python
def NF(w):
    x = ''
    for i in range(1, len(w)+1):
        #print(x)
        x = Mult(w[-i], x)
    return x
```

```python
def leftDescent(w):
    list = ['a', 'b','c']
    returnlist = []
    for s in list:
        bool = MultLen(s,w) < len(w)
        #print(m)
        #print(bool)
        if bool:
            returnlist.append(s)
    return returnlist
```

```python
def DescentPath(w):
    if len(w) == 1:
        return [[w], [w]]
    x = w[-1]
    elementList = [w[-1]]
    descentsList = []
    for i in range(1, len(w)):
        gens = ['a', 'b', 'c']
        descent = []
        for g in gens:
            if g == w[-i-1]:
                newx = Mult(g, x)
                if len(newx) < len(x):
                    descent.append(g)
                elementList.append(newx)
            else:
                if isLeftDescent(g, x):
                    descent.append(g)
        x = elementList[-1]
        #print(x)
        descentsList.append(descent)
    descentsList.append(leftDescent(x))
    return [elementList, descentsList]
    

```
